```python
import random
from time import sleep

class board:
    def __init__(self):
        self.board = [[' ', ' ', ' '], 
                      [' ', ' ', ' '],
                      [' ', ' ', ' ']]
        self.last_turn = 'cpu'
        self.win = False
        self.draw = False
        self.winner = 'No one'
        self.play()
        
    def cpu_move(self):
        moved = False
        
        while not moved:
            row = random.randint(0,2)
            col = random.randint(0,2)

            if self.board[row][col] == ' ':
                self.board[row][col] = 'O'
                self.last_turn = 'cpu'
                moved = True
            else:
                continue
                
    def human_move(self):
        moved = False
        
        while not moved:
            try:
                row = int(input("Enter row position: "))
                col = int(input("Enter column position: "))
            except:
                print('bad input. try again')
                continue
            
            if not 0 <= row <= 2 or not 0 <= col <= 2:
                print('out of bounds. try again')
                continue
                
            elif self.board[row][col] == ' ':
                self.board[row][col] = 'X'
                self.last_turn = 'human'
                moved = True
                
            else:
                print("space taken. pick another one")
                continue
                
                
    def print_board(self):
        for i, row in enumerate(self.board):
            print('', row[0], '|', row[1], '|', row[2], ' ')
            if i < 2:
                print('------------')
                
    def check_win(self):
       
        win_opts = [''.join(self.board[0]),
                    ''.join(self.board[1]),
                    ''.join(self.board[2]),
                    ''.join([item[0] for item in self.board]),
                    ''.join([item[1] for item in self.board]),
                    ''.join([item[2] for item in self.board]),
                    ''.join([self.board[i][i] for i in range(3)]),
                    ''.join([self.board[i][2-i] for i in range(3)])
                    ]
        
        if 'XXX' in win_opts:
            self.winner = 'You'
            self.win = True
            
        elif 'OOO' in win_opts:
            self.winner = 'CPU'
            self.win = True
            
        
    def check_draw(self):
        board_list = [item for sublist in self.board for item in sublist]

        if ' ' not in board_list:
            self.draw = True
                
    def play(self):
        while not self.win and not self.draw: 
            #print('')
            sleep(.5)
            self.print_board()
            
            if self.last_turn == 'cpu':
                self.human_move()
            else:
                self.cpu_move()
                
            self.check_draw()
            self.check_win()
                
        self.print_board()
        
        print('The winner is {}'.format(self.winner))
```
