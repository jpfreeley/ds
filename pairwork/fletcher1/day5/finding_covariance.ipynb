```python
import numpy as np
from random import randint
```

```python
mat = np.array([[randint(1,20) for _ in range(5)] for _ in range(20)])
mat
```

```python
means = np.mean(mat, axis=0)
means
```

```python
diff = np.array([i-means for i in mat])
```

```python
t = diff.transpose()
```

```python
cov = np.dot(diff,t)
```

```python
cov/4
```
