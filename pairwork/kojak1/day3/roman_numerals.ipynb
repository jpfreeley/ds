```python
def roman(num):
    final_string = []
    str_num = str(num)
    singles = {
                '1':'I',
                '2':'II',
                '3':'III',
                '4':'IV',
                '5':'V',
                '6':'VI',
                '7':'VII',
                '8':'VIII',
                '9':'IX',
                '0':''
    }
    tens = {
            '1':'X',
            '2':'XX',
            '3':'XXX',
            '4':'XL',
            '5':'L',
            '6':'LX',
            '7':'LXX',
            '8':'LXXX',
            '9':'XC',
            '0':''
    }
    hundreds = {
                '1':'C',
                '2':'CC',
                '3':'CCC',
                '4':'CD',
                '5':'D'
    }
    
    try:
        final_string.append(hundreds[str_num[-3]])
    except:
        pass
    
    try:
        final_string.append(tens[str_num[-2]])
    except:
        pass
    
    final_string.append(singles[str_num[-1]])
    
    return ''.join(final_string)
```
