```python
def roman(num):
    final_string = []
    str_num = str(num)
    singles = {
                '1':'I',
                '2':'II',
                '3':'III',
                '4':'IV',
                '5':'V',
                '6':'VI',
                '7':'VII',
                '8':'VIII',
                '9':'IX',
                '0':''
    }
    tens = {
            '1':'X',
            '2':'XX',
            '3':'XXX',
            '4':'XL',
            '5':'L',
            '6':'LX',
            '7':'LXX',
            '8':'LXXX',
            '9':'XC',
            '0':''
    }
    hundreds = {
                '1':'C',
                '2':'CC',
                '3':'CCC',
                '4':'CD',
                '5':'D'
    }
    
    try:
        final_string.append(hundreds[str_num[-3]])
    except:
        pass
    
    try:
        final_string.append(tens[str_num[-2]])
    except:
        pass
    
    final_string.append(singles[str_num[-1]])
    
    return ''.join(final_string)
```

```python
roman_list = []

for i in range(1,501):
    current_roman = roman(i)
    roman_list.append(current_roman)
    print('{} : {}'.format(i,current_roman))
```

```python
def from_romans(roman):
    final_value = 0
    values = {
                'I':1,
                'V':5,
                'X':10,
                'L':50,
                'C':100,
                'D':500
    }
    
    for i, letter in enumerate(roman[:]):
        #print(letter)
        final_value += values[letter]
        if i > 0:
            #check previous value
            if values[roman[i-1]] < values[letter]:
                final_value -= 2*values[roman[i-1]]
    
    return final_value
```

```python
from_romans('VIII')
```

```python
from_romans('CDLVII')
```

```python
from_romans('CCCXLVIII')
```

```python
for i in roman_list:
    print('{} : {}'.format(i, from_romans(i)))
```
